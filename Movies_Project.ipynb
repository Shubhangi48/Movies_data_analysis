```python
#dropping null from director and genres

df.dropna(how = 'any', subset=['director', 'genres'], inplace=True)
```

```python
#removing null by making the noll as 0 in keywords, production_companies

df['keywords']=df['keywords'].fillna(0)
df['production_companies']=df['production_companies'].fillna(0)
```

```python
#removing or rounding the decimal point to 2 in popularity column

df['popularity'] = df['popularity'].round(2)
```

```python
#inserting a profit column by subtracting budget and revenue column

df.insert(3, 'profit', df.revenue-df.budget)
```

```python
df
```

```python
#inserting another column as ROI by dividing profit by its budget

df.insert(4, 'ROI', df.profit/df.budget)
```
